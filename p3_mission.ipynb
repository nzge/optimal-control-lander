# Part III — Two-Phase Mission (Min-Time Ascent + Landing)

Indirect single shooting on Phase A (bang-bang minimum time) and Phase B (free-final-time LQR landing).

Modules: `mission.py`, `experiments.run_part3_mission`, `validation.validate_mission`

```python
import matplotlib.pyplot as plt
import numpy as np

import experiments as exp
import validation as val
from param import params

%matplotlib inline
```

```python
# Manifold M1: flat-ground touchdown (p_z = v_z = theta = 0)
p3 = exp.run_part3_mission(params, manifold="M1", verbose=True)
sol = p3["solution"]
val.print_mission_report(sol, params)
val.validate_mission("P3 M1", sol, params).summary()
```

```python
px = np.concatenate([sol.x_a[:, 0], sol.x_b[:, 0]])
pz = np.concatenate([sol.x_a[:, 1], sol.x_b[:, 1]])

fig, axes = plt.subplots(1, 2, figsize=(11, 4))
axes[0].plot(px, pz, "k-")
axes[0].set(xlabel=r"$p_x$", ylabel=r"$p_z$", title="Mission plane (M1)")
axes[1].plot(sol.t_a, sol.H_a, label=r"$H_A$")
axes[1].plot(sol.t_b, sol.H_b, label=r"$H_B$")
axes[1].axvline(sol.t1, color="gray", ls=":", label="handoff")
axes[1].axhline(0, color="k", lw=0.8)
axes[1].set(xlabel="t [s]", ylabel="H", title="Hamiltonian consistency")
axes[1].legend()
plt.tight_layout()
```
