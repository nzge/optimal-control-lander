# Part IV — Nonlinear Optimal Control Near Hover

Three required scenarios:
1. LQR on the **linearized** plant
2. The **same LQR** on the **nonlinear** plant
3. **Nonlinear PMP** (TPBVP) on the **nonlinear** plant

Modules: `nonlinear.py`, `experiments.run_part4_nonlinear`, `validation.validate_part4`

```python
import matplotlib.pyplot as plt
import numpy as np

import experiments as exp
import nonlinear as nl
import validation as val
from param import params

%matplotlib inline
```

```python
t = exp.default_time_grid(params)
p4 = exp.run_part4_nonlinear(params, t, theta_scale=1.0, verbose=True)
nl.print_part4_report(p4)
val.print_report(val.validate_part4(p4, params))
```

```python
sl, sn, sp = p4["scenario_linear"], p4["scenario_lqr_nl"], p4["scenario_pmp_nl"]

fig, ax = plt.subplots(figsize=(6, 6))
ax.plot(sl["x_hist"][:, 0], sl["x_hist"][:, 1], label=nl.SCENARIO_LABELS["linear"])
ax.plot(sn["x"][:, 0], sn["x"][:, 1], label=nl.SCENARIO_LABELS["lqr_nl"])
ax.plot(sp.x[:, 0], sp.x[:, 1], "--", label=nl.SCENARIO_LABELS["pmp_nl"])
ax.set(xlabel=r"$p_x$", ylabel=r"$p_z$", title="Mission plane")
ax.legend(fontsize=8)
ax.axis("equal")
plt.tight_layout()
```
